```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../JSON/Backup/totals.csv')
```

```python
feature = ['PTS', 'Team.Wins', 'FG', 'AST', 'DRB', 'TOV', 'PF']
```

```python
training = data[data['Share'] != 0]
```

```python
X = training[feature]
y = training['Share']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=42, n_jobs=-1, n_estimators=916, max_depth=16, min_samples_leaf=6, max_leaf_nodes=16, min_samples_split=3)
rf.fit(X_train, y_train)
rf.score(X_valid, y_valid)
```
